```python
from PIL import Image
import matplotlib.pyplot as plt
import requests
from io import BytesIO


url = "https://upload.wikimedia.org/wikipedia/commons/4/47/PNG_transparency_demonstration_1.png"
response = requests.get(url)
response.raise_for_status()
img = Image.open(BytesIO(response.content))


plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.imshow(img)
plt.axis('off')
plt.title('Original Image')

# Rotate image by 45 degrees
rotated = img.rotate(45, expand=True)  # expand=True to resize canvas to fit rotated image
plt.subplot(2, 3, 2)
plt.imshow(rotated)
plt.axis('off')
plt.title('Rotated 45')

# Resize image
resized = img.resize((img.width // 2, img.height // 2))
plt.subplot(2, 3, 3)
plt.imshow(resized)
plt.axis('off')
plt.title('Resized (50%)')

```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
from IPython.display import Audio

# Load an example audio file provided by librosa
audio_path = librosa.ex('trumpet')

# Load audio data
y, sr = librosa.load(audio_path)

# Display waveform
plt.figure(figsize=(10, 3))
librosa.display.waveshow(y, sr=sr)
plt.title('Audio Waveform')
plt.show()

# Play audio inline
Audio(audio_path)
```

```python
import cv2
from IPython.display import HTML
from base64 import b64encode
import matplotlib.pyplot as plt


!wget -q -O sample_video.mp4 https://filesamples.com/samples/video/mp4/sample_640x360.mp4

# Read the video using OpenCV
cap = cv2.VideoCapture('sample_video.mp4')
ret, frame = cap.read()
cap.release()

if ret:
    # Convert frame from BGR to RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.imshow(frame_rgb)
    plt.axis('off')
    plt.title('First Frame of Video')
    plt.show()
else:
    print("Failed to read video frame.")

# Embed and display video inline
mp4 = open('sample_video.mp4', 'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML(f"""
<video width=400 controls>
  <source src="{data_url}" type="video/mp4">
</video>
""")
```
