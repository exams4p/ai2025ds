```python

!pip install nltk
# Import library
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
# Download resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Download punkt_tab
# Sample text
text = "Python is a powerful language. It is used in data science, AI, and web development."
# Sentence tokenization
sentences = sent_tokenize(text)
print("Sentences:", sentences)
# Word tokenization
words = word_tokenize(text)
print("\nWords:", words)
# Removing stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words]
print("\nFiltered Words:", filtered_words)
# Frequency Distribution
fdist = FreqDist(filtered_words)
print("\nWord Frequency Distribution:")
print(fdist.most_common())
```

```python
# ==========================================
# ðŸ”— GRAPH DATA HANDLING USING NETWORKX
# ==========================================
# Install NetworkX
!pip install networkx
# Import library
import networkx as nx
import matplotlib.pyplot as plt
# Create a simple graph
G = nx.Graph()
# Add nodes
G.add_nodes_from(["A", "B", "C", "D"])
# Add edges
G.add_edges_from([("A", "B"), ("A", "C"), ("B", "D"), ("C", "D")])
# Display basic info
print("Nodes:", G.nodes())
print("Edges:", G.edges())
# Draw the graph
plt.figure(figsize=(5, 4))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=1500, font_size=12)
plt.title("Simple Graph using NetworkX")
plt.show()
```

```python
# ==========================================
# ðŸŒ SPATIAL DATA HANDLING USING GEOPANDAS
# ==========================================
# Install Geopandas
!pip install geopandas
# Import library
import geopandas as gpd
from shapely.geometry import Point, Polygon
# Create sample spatial data
data = {
    'City': ['Mumbai', 'Delhi', 'Bangalore'],
    'Latitude': [19.0760, 28.6139, 12.9716],
    'Longitude': [72.8777, 77.2090, 77.5946]
}
# Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data['Longitude'], data['Latitude']))
# Display GeoDataFrame
print(gdf)
# Plot the points
gdf.plot(color='green', markersize=80)
plt.title("Cities plotted using GeoPandas")
plt.show()
```
