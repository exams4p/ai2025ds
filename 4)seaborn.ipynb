```python
# -----------------------------------------
# ðŸ“Š Import Required Libraries
# -----------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load an inbuilt dataset (for simplicity)
iris = sns.load_dataset('iris')
```

```python
# Shows how sepal length and width relate to different species
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set2')
plt.title("Feature Relationship for Iris Classification")
plt.show()
```

```python
# Example data
epochs = np.arange(1, 6)
train_acc = [60, 70, 78, 85, 90]
val_acc = [58, 67, 75, 80, 87]

data = pd.DataFrame({'Epoch': epochs, 'Train Accuracy': train_acc, 'Validation Accuracy': val_acc})
sns.lineplot(data=data[['Epoch', 'Train Accuracy']], x='Epoch', y='Train Accuracy', label='Train')
sns.lineplot(data=data[['Epoch', 'Validation Accuracy']], x='Epoch', y='Validation Accuracy', label='Validation')
plt.title("Model Accuracy Over Epochs")
plt.show()
```

```python
sns.histplot(iris['petal_length'], bins=20, kde=True, color='orchid')
plt.title("Feature Distribution â€“ Petal Length")
plt.show()
```

```python
sns.boxplot(data=iris, x='species', y='sepal_width', palette='coolwarm')
plt.title("Outlier Detection by Species")
plt.show()
```

```python
corr = iris.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Feature Correlation Heatmap")
plt.show()
```
