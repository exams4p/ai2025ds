```python
# -----------------------------------------
# ðŸ“Š Import Required Libraries
# -----------------------------------------
import matplotlib.pyplot as plt
import numpy as np

# For consistent style
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
# Example data
epochs = [1, 2, 3, 4, 5]
train_acc = [60, 70, 78, 85, 90]
val_acc = [58, 67, 75, 80, 87]

# Plot
plt.plot(epochs, train_acc, marker='o', label='Training Accuracy')
plt.plot(epochs, val_acc, marker='s', label='Validation Accuracy')
plt.title("Model Accuracy Over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.show()
```

```python
# Example data
classes = ['Cat', 'Dog', 'Horse', 'Bird']
counts = [200, 180, 150, 220]

# Plot
plt.bar(classes, counts, color='lightcoral')
plt.title("Class Distribution in Dataset")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.show()
```

```python
# Generate random cluster data
x1 = np.random.randn(50) + 2
y1 = np.random.randn(50) + 2
x2 = np.random.randn(50) - 2
y2 = np.random.randn(50) - 2

# Plot
plt.scatter(x1, y1, color='blue', label='Cluster 1')
plt.scatter(x2, y2, color='green', label='Cluster 2')
plt.title("Cluster Visualization using Scatter Plot")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# Example: distribution of pixel intensity or feature values
data = np.random.randn(1000)

plt.hist(data, bins=20, color='orchid', edgecolor='black')
plt.title("Feature Value Distribution")
plt.xlabel("Feature Value")
plt.ylabel("Frequency")
plt.show()
```

```python
labels = ['Healthy', 'Diabetic', 'Hypertensive']
sizes = [60, 25, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'lightgreen'])
plt.title("Medical Dataset Composition")
plt.show()
```
